```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the dataset
data = pd.read_csv('netflix_titles.csv')

# View first few rows
print(data.head())
```

```python
print(data.info())
print(data.shape)
```

```python
print(data.isnull().sum())
```

```python
data.dropna(subset=['director', 'cast', 'country'], inplace=True)

```

```python
data['country'] = data['country'].fillna('Unknown')
data.drop_duplicates(inplace=True)

```

```python
# Remove spaces
data['date_added'] = data['date_added'].str.strip()

# Convert safely
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
```

```python
# Create a new column of lists
data['genres'] = data['listed_in'].apply(lambda x: x.split(', '))
```

```python
type_counts = data['type'].value_counts()

sns.barplot(x=type_counts.index, y=type_counts.values, palette='Set2')
plt.title('Distribution of Content by Type')
plt.show()
```

```python
all_genres = sum(data['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)

sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='Set3')
plt.title('Most Common Genres')
plt.show()
```

```python
data['title_length'] = data['title'].apply(len)
```

```python
data['num_genres'] = data['listed_in'].apply(lambda x: len(x.split(',')))
```

```python
def map_rating(r):
    if r in ['TV-MA', 'R', 'NC-17']:
        return 'Adult'
    elif r in ['PG-13', 'TV-14']:
        return 'Teens'
    elif r in ['PG', 'TV-PG', 'TV-Y7', 'TV-Y']:
        return 'Kids'
    elif r in ['G', 'TV-G']:
        return 'All Ages'
    else:
        return 'Unrated'

data['age_group'] = data['rating'].apply(map_rating)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Use engineered features
features = data[['title_length', 'num_genres']]
target = data['type']

# Encode target
le = LabelEncoder()
y = le.fit_transform(target)

X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Clean the title for comparison
    title = title.strip()
    
    matches = data[data['title'].str.lower() == title.lower()]
    
    if matches.empty:
        print(f"No exact match found for title: {title}")
        similar = data[data['title'].str.contains(title.split()[0], case=False, na=False)]
        if not similar.empty:
            print("Did you mean one of these?")
            print(similar['title'].unique())
        return []
    
    idx = matches.index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:6]
    recs = [data['title'].iloc[i[0]] for i in sim_scores]
    return recs
```
